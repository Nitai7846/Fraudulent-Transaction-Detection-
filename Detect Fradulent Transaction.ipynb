```python
#import data preprocessing libraries
import pandas as pd
import numpy as np
import sklearn
```

```python
data = pd.read_csv("/Users/nitaipremalshah/Downloads/Fraud.csv")
```

```python
#get a basic idea about the data
data.describe()
```

```python
#save a original copy of the data
data_new = data.copy()
```

```python
# create a new columnn and initialize it with NaN values
data_new['type1'] = np.nan
```

```python
data_new
```

```python
#Fill the new columns
#CC denotes customer to customer transaction, and other so on
data_new.loc[data.nameOrig.str.contains('C')  & data.nameDest.str.contains('C'), "type1"] = "CC"
data_new.loc[data.nameOrig.str.contains('C')  & data.nameDest.str.contains('M'), "type1"] = "CM"
data_new.loc[data.nameOrig.str.contains('M')  & data.nameDest.str.contains('C'), "type1"] = "MC"
data_new.loc[data.nameOrig.str.contains('M')  & data.nameDest.str.contains('M'), "type1"] = "MM"
```

```python
#create fraud and valid columns
fraud = data_new[data_new["isFraud"] == 1]
valid = data_new[data_new["isFraud"] == 0]
```

```python
#IT appears customer to customer transactions contain ALL the fradulent transactions
print("Fraud Transactions by type1 : \n",  fraud.type1.value_counts())
print("Valid Transactions by type1 : \n",  valid.type1.value_counts())
```

```python
#drop fraud, valid and type 1 columns as they are no longer needed
fraud = fraud.drop('type1', 1)
valid = valid.drop('type1',1)
data_new = data_new.drop('type1',1)
```

```python
#It appears Cash Out and Transfer are the only types in which fraudulent transaction take place
print("Fraud Transactions by type : \n", fraud.type.value_counts())
print("Valid Transactions by type : \n", valid.type.value_counts())
```

```python
#Create a new dataframe with only these types
valid = valid[(valid["type"] == "CASH_OUT")| (valid["type"] == "TRANSFER")]
data_new = data_new[(data_new["type"] == "CASH_OUT") | (data_new["type"] == "TRANSFER")]
```

```python
#Names do not really hold a co-relation with fraudulent transactions hence they can be dropped
names = ["nameOrig","nameDest"]
fraud = fraud.drop(names, 1)
valid = valid.drop(names,1)
data_new = data_new.drop(names,1)
```

```python
data_new
```

```python
#isFlaggedFraud has only 16 fraudulent transactions compared to ~ 6mn entries. This column can be hence removed
fraud = fraud.drop("isFlaggedFraud",1)
valid = valid.drop("isFlaggedFraud",1)
data_new = data_new.drop("isFlaggedFraud",1)
```

```python
data_new
```

```python
#create a new feature column where we divide step by 24 to see the effect time of the day has on fradulent transactions

dataset1 = data_new.copy()


# adding feature HourOfDay to Dataset1 
dataset1["HourOfDay"] = np.nan # initializing feature column
dataset1.HourOfDay = data_new.step % 24


print("Head of dataset1: \n", pd.DataFrame.head(dataset1))
```

```python
dataset = dataset1.copy() 
```

```python
#perform label encoding for categorical variables, in this case, the type column
dataset = pd.get_dummies(dataset,prefix=['type'])
```

```python
dataset
```

```python
# create the feature and target variable columns
X = dataset.drop("isFraud",1)
y = dataset.isFraud
```

```python
#split these into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
#Use scaling in order to ensure numerically larger values do not influence the model, and cause biases to creep in
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Try a basic logistic regression
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression()
model_logistic.fit(X_train, y_train)
y_pred = model_logistic.predict(X_test)

```

```python
#import the various metrics for classification
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score, roc_curve, auc,\
precision_score
```

```python
#define a funtion to display classification metrics
def classification_results(arr1, arr2):
    
    print(classification_report(arr1, arr2))
    print("\n")
    
    print(confusion_matrix(arr1, arr2))
    print("\n")
    
    print(accuracy_score(arr1, arr2))
    print("\n")
    
   
    
    
    
```

```python
#get results for logistic regression
classification_results(y_pred, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
```

```python
#fit a random forest model and generate predictions on test set
model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict(X_test)
classification_results(y_pred_rf, y_test)
```

```python
#fit a XGBClassifier model
model_xgb = XGBClassifier()
model_xgb.fit(X_train, y_train)
```

```python
#generate predictions on test set
y_pred_xgb = model_xgb.predict(X_test)
classification_results(y_pred_xgb, y_test)
```

```python
#import neccesary libraries for a ANN 
import tensorflow
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#create a ann for binary classification
model_ann = Sequential()
model_ann.add(Dense(512, activation='relu', input_dim=9))
model_ann.add(Dense(512, activation='relu'))
model_ann.add(Dense(1, activation='sigmoid'))
model_ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model_ann.fit(X_train, y_train, epochs=10, batch_size=100)
```

```python
#Generate predictions on test set
y_pred_ann = model_ann.predict(X_test)
```

```python
#round values in order for binary classificiation metrics 
y_pred_ann = y_pred_ann.round()
```

```python
#metrics for ann
classification_results(y_pred_ann, y_test)
```

```python
#RESULT DISCUSSION
#FOUR different models were tried. 
#Let us take the confusion matrix in discussion 
#The best performing models are the XGBoost and RandomForestClassifier, followed by the ANN and logistic regression
#Though the XGBoost has a lower number of False Positives, the Random Forest Regressor has a lower number of false negatives.
#In the case of fradulent transactions it would be ok to send fake alerts for fraudulent transactions, instead of missing out on genuine fradulent transactions. 
#Hence we can say that in the case of this dataset the RF classifier perfroms the best, the XGboost classifier the second best, the ANN the third best followed by the logistic regressor at last place.
```
